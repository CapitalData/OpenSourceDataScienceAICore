##### **What Is SQL?**
- SQL stands for Structured Query Language.
- It's the standard language for interacting with databases—used to retrieve, insert, update, and delete data.
- Communicates with relational databases.
- Whether you're a data scientist, backend developer, or business analyst, knowing SQL is a game changer.
- Almost every big tech company uses SQL to manage and organize data be it Uber, Netflix, Airbnb, Facebook, Google, LinkedIn, or any other, SQL is everywhere.



#####**What is Relational Database?**
-  A type of database that stores data in tables (also called relations) that are linked to each other through relationships.

image.png

#### **Key Concepts**
- Tables
  - Structured collections of rows and columns

- Rows
  - Individual records (like a spreadsheet row)

- Columns
  - Attributes or fields (like “name”, “email”)

- Primary Key
  - A unique identifier for each row

- Foreign Key
 - A reference to a primary key in another table (creates a relationship)



#### **SQL Processing**

image.png

- **SQL parsing:**
  - SQL engine checks for syntax errors
  - Converts the query into an internal form called a parse tree.
- **SQL Optimization:**
  - The query optimizer finds the most efficient execution plan
  - It decides:
   - Which indexes to use
   - Join order (if applicable)
   - Whether to use sequential or indexed scan
- **Row Source Generation:**
  - Row Source Generation is the phase where the SQL engine converts an execution plan into a tree of iterators (aka row sources), each producing rows of data.
 - Think of it as building a pipeline of steps, where each step:
   - Takes input rows
   - Processes them (filtering, joining, sorting, etc.)
   - Passes them to the next step
- **Execution:**
  - The query runs:
   - Data is retrieved, filtered, sorted, or aggregated
   - Results are returned to you



### **SQL Commands**
**1. DDL(Data Definition Language)**
 - Used to define and manage database structure (tables, schemas, etc.)
 - CREATE
    - Create tables/databases
    ```CREATE TABLE users (id INT, name VARCHAR(50));```
 - ALTER
   - Modify existing table
   ```ALTER TABLE users ADD email VARCHAR(100);```
 - DROP
   - Delete table/database
    ``` DROP TABLE users;```
 - TRUNCATE
   - Delete all rows (fast)
   ```TRUNCATE TABLE users;```

image.png

**2. DML (Data Manipulation Language)**
- Used to manage data inside the tables.
- Commands:
 - SELECT
   - Query data
 ``` SELECT * FROM users;```
 - INSERT
   - Add new records
     ``` INSERT INTO users (id, name) VALUES (1, 'Alice');```
 - UPDATE
   - Modify existing data
   ``` UPDATE users SET name = 'Bob' WHERE id = 1;```
 - DELETE
   - Remove data
    ``` DELETE FROM users WHERE id = 1;```

**3. DCL (Data Control Language)**
 - Used for access control.
 - Commands:
  - GRANT
    - Give user privileges
    ``` GRANT SELECT ON users TO 'user1';```
  - REVOKE
    - Remove user privileges
    ``` REVOKE SELECT ON users FROM 'user1';```

**4. TCL (Transaction Control Language)**
 - Used to control transactions.
 - Commands:
  - BEGIN
    - Starts a transaction
    ``` BEGIN;```
  - COMMIT
    - Saves all changes
    ``` COMMIT;```
  - ROLLBACK
    - Reverts changes if error occurs
    ``` ROLLBACK;```
  - SAVEPOINT
   - Sets a point to roll back to
    ``` SAVEPOINT before_update;```

**5. DQL(Data Query Language)**
- DQL statements are used for performing queries on the data within schema objects.
- It consists primarily of the SELECT statement.
```
SELECT first_name, last_name, hire_date
FROM employees
WHERE department = 'Sales'
ORDER BY hire_date DESC;
```

###**Functions in SQL**

- SQL is not just about reading rows , it’s about transforming and analyzing data right inside the query! That’s where functions come in.
- Want to count sales, format names, calculate discounts, or manipulate dates? SQL functions let you do all that in one step , no spreadsheet needed, no code outside the database.
- Funtions are classified into following groups according it's usecase:
  - **Aggregate Functions**:
    - Summarize data
    - These functions perform calculations on a set of values and return a single result.
    - Examples: SUM, AVG, COUNT, MIN, MAX
       ```
       SELECT SUM(salary) AS Total_salary,Max(salary) AS Highest_salary,
       MIN(salary) AS Lowest_salary,AVG(salary) AS Avg_salary'
       COUNT(salary) AS Count_Of_Salary
       from employee_salary;
       ```
  - **Scalar Functions** :
     -  Work on single values
     - These are classified as:
      - **String Functions** : Clean and format text
      - **Date Functions**: Calculate time differences
      - **Math Functions** : Rounding, modulus, etc.

#### **SQL Clauses**

SQL clauses are keywords used to filter, sort, or structure your query.
- **SELECT Clause**
  - Used to retrieve data from one or more tables.
- **FROM Clause**
  - Specifies the table(s) to retrieve data from.  
- **WHERE Clause**
  - Filters rows based on conditions.
- **WITH Clause**
  - The WITH clause defines a temporary named result set (called a Common Table Expression or CTE) that can be referenced in the main query.
- **HAVING Clause**
  - Filters groups after GROUP BY (like WHERE but for groups).
- **ORDER By Clause**
  - Sorts results in ascending or descending order.
- **Group By Clause**
  - Groups rows that have the same values in specified columns.
- **LIMIT/TOP Clause**
  - Restricts the number of rows returned.
- **Distinct Clause**
  -  Remove duplicate rows from your query results.
- **JOIN**
  - Combines rows from two or more tables.
- **UNION / UNION ALL**
  - Combines results from multiple SELECT queries.
- **Aliases**
  - An alias is a temporary name you give to a table or column in your SQL query to make results or logic easier to read.

**How To Use  GROUP BY and HAVING Clause?**
- **GROUP BY**  is used to group rows that have the same values in specified columns.
- Often used with aggregate functions like SUM(), COUNT(), AVG(), etc.
- **HAVING** Clause is used after GROUP BY to filter groups based on a condition.
- Like WHERE, but for groups, not individual rows.
- Example 1 : Shows only departments with more than 5 employees
```
 SELECT department, COUNT(*) AS total_employees
FROM employees
GROUP BY department
HAVING COUNT(*) > 5;
 -- WHERE COUNT(*) > 5 will through error.
```
- Example 2: Returns customers who spent more than 1000.
 ```
  SELECT customer_id, SUM(amount) AS total_spent
FROM orders
GROUP BY customer_id
HAVING SUM(amount) > 1000;
```


**Key Difference Between WHERE And HAVING**

Clause|Filters On|	Used With| Example
-------|--------|-----------|---------
WHERE|	Individual rows	|Before grouping|```SELECT * FROM orders WHERE status = 'delivered';```
HAVING|	Aggregated groups|	After grouping|```SELECT department, COUNT(*) FROM employees GROUP BY department HAVING COUNT(*) > 5;```

#### **JOIN in SQL**

SQL JOIN is used to combine rows from two or more tables based on a related column between them.

image.png

#### **Join Type**
- **INNER JOIN**
  - Only matched rows
  - Example:  Only students who are enrolled.
   ```
     SELECT a.name, b.course
     FROM students a
     INNER JOIN enrollments b ON a.id = b.student_id;
     ```

- **LEFT JOIN**
  - All from left + matched from right
  - Example: All students, even if they’re not enrolled.
  ```
    SELECT a.name, b.course
    FROM students a
    LEFT JOIN enrollments b ON a.id = b.student_id;
    ```

- **RIGHT JOIN**
  - All from right + matched from left
  - Example: Display all students ,even if student info is missing.
   ```
    SELECT a.name, b.course
    FROM students a
    RIGHT JOIN enrollments b ON a.id = b.student_id;
    ```
- **FULL OUTER JOIN**
  - All rows from both, matched or not
  - MySQL does not support FULL OUTER JOIN directly — but you can simulate it using UNION:
  
  ```
   SELECT a.*, b.*
   FROM table_a a
   LEFT JOIN table_b b ON a.id = b.id

   UNION

   SELECT a.*, b.*
   FROM table_a a
   RIGHT JOIN table_b b ON a.id = b.id;
  ```
- **CROSS JOIN**
  - All possible combinations (no condition)

#### **Subqueries**
- A subquery is a query within a query(Nested Query).
- **Syntax**:
```
 SELECT column_name
 FROM table_name
 WHERE column_name operator (SELECT column_name FROM table_name WHERE condition);
```
- **Operator:** Can be =, >, <, IN, NOT IN, EXISTS, ANY, or ALL.
- It appears within the WHERE or HAVING clause of another
SQL statement.
- **Example:Using a Single-Row Subquery in a HAVING Clause**
 - HAVING clause is used to filter groups of rows.
 - This allows you to filter groups of rows based on the result returned by your subquery.
 ```
 SELECT Department, AVG(Salary) AS AvgDeptSalary
FROM Employees
GROUP BY Department
HAVING AVG(Salary) > (SELECT AVG(Salary) FROM Employees);
```
- **Example: Subqueries in a FROM clause**
 ```
 SELECT E.Name, E.Salary, AvgSalary.Avg_Salary
FROM Employees E,
     (SELECT AVG(Salary) AS Avg_Salary FROM Employees) AvgSalary
WHERE E.Salary > AvgSalary.Avg_Salary;

#### **UNION vs UNION ALL**
**UNION**
- Combines results from two SELECT queries
- Removes duplicates (performs DISTINCT)
- Slightly slower due to duplicate-checking
- Syntax:
 ```
  SELECT SELECT column1, column2
  FROM table1
  UNION
  SELECT column1, column2
  FROM table2;
  ```

**UNION ALL**
- Combines all rows from both queries
- Keeps duplicates
- Faster than UNION
- Syntax:
 ```
  SELECT SELECT column1, column2
  FROM table1
  UNION ALL
  SELECT column1, column2
  FROM table2;
  ```

###**Indexes & Performance Optimization**

**What is Index?**
- An index in SQL is like a lookup table that helps the database find data faster, just like a book’s index helps you find topics without reading every page.
-  It is a shortcut for the database engine to jump to the rows it needs.
- Without index: the DB scans every row (called Full Table Scan).
- With index : It jumps directly to matching rows.



image.png

**Why To Use Indexes?**
- Faster queries with WHERE, JOIN, ORDER BY, GROUP BY
- Efficient data access in large datasets
- Improves performance in data analysis and reporting

**When Not to Use Indexes**
- Columns with high duplicate values (like gender or status)
- Tables that are small — indexes may not help much
- Columns that are frequently updated — indexing can slow down INSERT/UPDATE/DELETE



**Tools for Analyzing Performance**
- EXPLAIN or EXPLAIN ANALYZE — shows query plan and if index is used

- SHOW INDEXES FROM table_name — see existing indexes



**Performance Optimization Tips**

**Goal**	| **Strategy**
----------|--------------
Fast filtering	|Add index on WHERE columns
Speed up joins	| Index on join keys
Faster sorting/grouping |	Index on ORDER BY / GROUP BY
Reduce scan cost	| Use EXPLAIN to debug slow queries
Avoid unused indexes	|Too many indexes can slow writes

- **id: 1** - This is a unique identifier for this particular step in the query execution plan. Simple queries usually have a single id of 1. More complex queries might have multiple steps, each with its own id.

- **type:** ref - This is one of the most important columns. It describes how MySQL is accessing the table. ref means that MySQL is using an index to quickly find the rows that match a specific value in a WHERE clause. This is generally an efficient access method.

- **possible_keys:** customer_id - This lists the indexes that MySQL could have used to find the relevant rows.

- **key_len:** 5 - This is the length of the key that MySQL used from the index. It can give you an idea of how much of the index is being used.

- **ref:** const - This shows the value that MySQL is using to look up rows in the index. const means that the value is a constant, like a number or a string that you've provided in the WHERE clause (e.g., WHERE customer_id = 1).

- **rows**: 3 - This is MySQL's estimate of the number of rows it needs to examine to find the result. A lower number is generally better.

- **filtered**: 100.0 - This gives an estimate of how many rows examined by this step will be filtered out by subsequent steps in the plan. 100.0 means that all rows examined by this step are likely part of the final result.

- **Extra**: None - This column provides additional information. In this case, None means there's no extra information to report.

### **Triggers and Event Scheduling**


##### **Trigger**
- A special type of stored procedure in SQL that is automatically executed (or "triggered") when a specified event occurs in a table.
- Common events that activate triggers include INSERT,UPDATE and DELETE.
- **Syntax:**
```
  CREATE TRIGGER trigger_name
  [BEFORE | AFTER] [INSERT | UPDATE | DELETE]
  ON table_name
  FOR EACH ROW
  BEGIN
    -- SQL statements
  END;
  ```
- **Types of Triggers**
 - Triggers can be classified based on:

   - Timing:
    - BEFORE Triggers – Execute before the operation (INSERT, UPDATE, DELETE).

    - AFTER Triggers – Execute after the operation.

  -  Triggering Event

     - INSERT Triggers – Fires when a new row is inserted.

    - UPDATE Triggers – Fires when an existing row is updated.

    - DELETE Triggers – Fires when a row is deleted.
  - Scope of Execution
    - Row-Level Trigger
     - Executes once per row affected by INSERT, UPDATE, or DELETE.
     - Ideal for logging individual changes.
    - Statement-Level Trigger
     - Executes once per statement, regardless of the number of rows affected.
     - Ideal for tracking table modifications.  

#### **Event Scheduling**
- SQL Events are scheduled tasks that execute automatically at specified times or intervals, making them useful for automating maintenance, data updates, and background processes.
- They work similarly to cron jobs in Linux or scheduled tasks in Windows.
- **Syntax:**
```
CREATE EVENT [IF NOT EXIST] event_name
ON SCHEDULE schedule
DO
event_body;
```
- **Two types of events:**
  1. One time
  2. Recurrent

**5. Create Materialized View**
- Although Materialized view is not supported by MySQL , we can impliment with MySQl using Refresh Function.
- Refresh function is used to update the data in that view to reflect changes made in the base tables.
- Let's dive in with example.

````python
# First create table and name it as materialized view
cursor.execute("DROP TABLE IF EXISTS materialized_employee_view")

# Define materialized view
cursor.execute("""
CREATE TABLE materialized_employee_view AS
SELECT e.emp_id, e.name, d.dept_name, e.salary
FROM employees e
JOIN departments d ON e.dept_id = d.dept_id
""")
````

````python
# define refresh function
def refresh_materialized_view():
    #clears the old snapshot with TRUNCATE
    cursor.execute("TRUNCATE materialized_employee_view")
    cursor.execute("""
    INSERT INTO materialized_employee_view
    SELECT e.emp_id, e.name, d.dept_name, e.salary
    FROM employees e
    JOIN departments d ON e.dept_id = d.dept_id
    """)
    conn.commit()
````

````python
cursor.execute("UPDATE employees SET salary = 100000 WHERE emp_id = 1")
conn.commit()

# View (Dynamic)
cursor.execute("SELECT * FROM employee_view WHERE emp_id = 1")
print("VIEW:", cursor.fetchall())

# Materialized View (Old until refresh)
cursor.execute("SELECT * FROM materialized_employee_view WHERE emp_id = 1")
print("MATERIALIZED VIEW:", cursor.fetchall())

# Refresh + Query
refresh_materialized_view()
cursor.execute("SELECT * FROM materialized_employee_view WHERE emp_id = 1")
print("MATERIALIZED VIEW (After Refresh):", cursor.fetchall())
````

#### **Where to Use View And Materialized View?**

Use Case	|Use a VIEW	|Use a MATERIALIZED VIEW (or simulated one in MySQL)
----------|------------|------------------------------------
Data should always reflect the latest changes|	✅ Yes|	❌ No
You want fast reads on heavy queries|❌ No|	✅ Yes
Base data changes frequently and performance is not a concern|✅ Yes	|❌ No
You want to avoid running expensive joins or aggregations every time	|❌ No	|✅ Yes
Data is analyzed periodically (e.g. daily reports)	|❌ Maybe|	✅ Yes


#### **Refresh Strategy:**
If you're going with materialized view (simulated), you need to decide how and when to refresh:
- Manual Refresh
  - Best for dev/testing or when updates are infrequent.
  - ```refresh_materialized_view()``` Defined in code with materialized view.

- Auto Refresh
  - Recommended for production
  - Using MySQL Event Scheduler:
   ```
    -- Enable event scheduler
    SET GLOBAL event_scheduler = ON;

   -- Schedule a refresh every day at midnight
   CREATE EVENT refresh_mview_event
   ON SCHEDULE EVERY 1 DAY
   STARTS CURRENT_DATE + INTERVAL 1 DAY
     DO
    CALL refresh_materialized_view();  -- or use raw SQL refresh logic
  ```  

#### **Window Functions: ROW_NUMBER(), SUM() OVER, RANK()**

- Window functions use values from multiple rows to produce values for each row separately.
- In other words, they perform calculations on the set of rows related to the current row.

#### **When to Use Window Functions:**
- Ranking top-N items per group
- Running totals / moving averages
- Comparing current row to previous/next (time series)
- Gap analysis
- Percentiles and cumulative distributions

**Syntax**:

 ```
  SELECT
  window_function() OVER(
          PARTITION BY partition_expression
         ORDER BY order_expression
         window_frame_extent
     ) AS window_column_alias
    FROM table_name;
  ```
  - window_function()	The window function you want to apply (e.g., SUM, ROW_NUMBER)
  - OVER()	Defines the window of rows to perform the calculation on
  - PARTITION BY	(Optional) Groups rows for separate window calculations (like GROUP BY)
  - ORDER BY	(Optional) Orders rows within each partition
  - ROWS BETWEEN	(Optional) Specifies a range of rows (e.g., previous 1 row to next 1 row)

####**CTE (Common Table Expressions)**

- **A Common Table Expression (CTE) is the result set of a query** which exists temporarily and for use only within the context of a larger query.
- **Syntax**

 ```
 WITH cte_name (column1, column2, ...)
 AS (
    -- Your Query Here
    SELECT column1, column2, ...
    FROM table_name
    WHERE condition
 )
 -- Use the CTE in your main query
  SELECT *
 FROM cte_name;
```
