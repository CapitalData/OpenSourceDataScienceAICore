### **SQL's Role in Relational Databases**
- Relational databases store data in tables with predefined schemas (rows and columns), ensuring data consistency, integrity, and relationships between entities.

#### **SQL plays a critical role in:**

**1. Data Definition 🏗️**

 - SQL allows users to define tables, columns, relationships, and constraints in a structured format.
 - Example:
 ```
 CREATE TABLE employees (
    id INT PRIMARY KEY,
    name VARCHAR(100),
    department VARCHAR(50),
    salary DECIMAL(10,2)
);
```

**2. Data Manipulation 🔄**
 - SQL enables inserting, updating, deleting, and retrieving structured data.
 - Example:
 ```
 INSERT INTO employees (id, name, department, salary)
VALUES (1, 'Alice', 'HR', 60000.00);
```

**3.Data Querying 🔍**
 - SQL allows users to extract meaningful insights from databases using queries.
 - Example:
 ```
 SELECT name, salary FROM employees WHERE department = 'HR';
 ```

**4. Data Integrity & Security 🔐**
  - SQL ensures data accuracy and security through constraints (e.g., PRIMARY KEY, FOREIGN KEY, UNIQUE, CHECK) and user access controls.
  - Example:
  ```
  ALTER TABLE employees ADD CONSTRAINT unique_name UNIQUE (name);
```


### **Interacting with Structured Data Using SQL**

**1. Retrieving Data (SELECT Queries)**
  - Extract specific data based on conditions.
  - Example:
   ```
   SELECT name, department FROM employees WHERE salary > 50000;
  ```

**2. Filtering & Sorting**
  - SQL allows sorting (ORDER BY) and filtering (WHERE, HAVING).  
  - Example:
  ```
  SELECT * FROM employees ORDER BY salary DESC;
 ```

**3. Joining Multiple Tables**
  - SQL enables combining related data across multiple tables using JOINs.
  - Example:
  ```
  SELECT employees.name, departments.department_name
  FROM employees
  JOIN departments ON employees.department_id = departments.id;
  ```

**4. Aggregating Data**
  - SQL helps in summarizing large datasets using functions like COUNT, SUM, AVG, MAX, MIN.  
  - Example:
  ```
  SELECT department, AVG(salary) AS avg_salary FROM employees GROUP BY department;
 ```

### **Basic SQL Architecture**

#### **Fundamental components of SQL databases**

**What is RDBMS?**
- Relational Database Management System,RDBMS is the basis for SQL, and for all modern database systems like MS SQL Server, IBM DB2, Oracle, MySQL, and Microsoft Access.

**1. Databases**
- A database is an organized collection of data stored electronically.
- It serves as a container for tables and other database objects.
- Examples: MySQL Database, PostgreSQL Database, Oracle Database
- A database contains multiple tables that store structured data.


**2. Tables**
- A table is the fundamental structure in an SQL database where data is stored in a structured format.
- Consists of rows (records) and columns (attributes).
- Each table has a name and a defined schema (structure).
```
CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    name VARCHAR(50),
    department VARCHAR(50),
    salary DECIMAL(10,2)
);
```

- A row contains values for each column in the table.
- Each row is uniquely identified, often by a Primary Key.

- Columns define the structure of a table by specifying the type of data that can be stored.
- Each column has a name and a data type.

**What is a NULL Value?**
- A NULL value in a table is a value in a field that appears to be blank, which means a field with a NULL value is a field with no value.
- It is very important to understand that a NULL value is different than a zero value or a field that contains spaces.


**SQL Constraints**
- Constraints are the rules enforced on data columns on a table.
- Used to limit the type of data that can go into a table.
- Ensures the accuracy and reliability of the data in the database.
- It can either be column level or table level.
- Column level constraints are applied only to one column whereas, table level constraints are applied to the entire table.
- Different Contraints
 1. NOT NULL Constraint
   - Ensures that a column cannot have a NULL value.
 2. DEFAULT Constraint  
   - Provides a default value for a column when none is specified.
 3. UNIQUE Key
   - Ensures that all the values in a column are different.
 4. PRIMARY Key
   - Uniquely identifies each row/record in a database table
   - It contains only a unique value.
   - It can not be null.
   - One table can have only one Primary Key.
   - A primary key length cannot be more than 900 bytes.   
 5. FOREIGN Key
   - Uniquely identifies a row/record in any another database table.
   - It is a column in one table that matches a Primary Key in another table, allowing the two tables to be connected together.
 6.	CHECK Constraint
    - Ensures that all values in a column satisfy certain conditions.
 7.	INDEX Constraint
    - Used to create and retrieve data from the database very quickly.   


**Data Integrity**
- Entity Integrity
  - This ensures that there are no duplicate rows in a table.

- Domain Integrity
   -  Enforces valid entries for a given column by restricting the type, the format, or the range of values.

- Referential integrity
  - Rows cannot be deleted, which are used by other records.

- User-Defined Integrity
  -  Enforces some specific business rules that do not fall into entity, domain or referential integrity.

#### **SQL Workflow**

- When you are executing an SQL command for any RDBMS, the system determines the best way to carry out your request and SQL engine figures out how to interpret the task.
- There are various components included in the process.
- These components are Query Dispatcher, Optimization
Engines, Classic Query Engine and SQL Query Engine.

image.png



- **SQL processing** is the **parsing**, **optimization, row source generation, and execution** of a SQL statement.
- **SQL parsing:**
  - SQL engine checks for syntax errors
  - Converts the query into an internal form called a parse tree.
- **SQL Optimization:**
  - The query optimizer finds the most efficient execution plan
  - It decides:
   - Which indexes to use
   - Join order (if applicable)
   - Whether to use sequential or indexed scan
- **Row Source Generation:**
  - Row Source Generation is the phase where the SQL engine converts an execution plan into a tree of iterators (aka row sources), each producing rows of data.
 - Think of it as building a pipeline of steps, where each step:
   - Takes input rows
   - Processes them (filtering, joining, sorting, etc.)
   - Passes them to the next step
- **Execution:**
  - The query runs:
   - Data is retrieved, filtered, sorted, or aggregated
   - Results are returned to you

2. **Statements**:

5. **Predicates**:
  - Predicates are essential for specifying conditions within SQL operations.

### **Some Key terms of SQL**

1. **Data Manipulation**:
  - The Data Manipulation Language (DML), focuses on operations that involve
     - Adding,
     - Updating,
     - Deleting data.
  - DML statements are pivotal for modifying the content of a database, ensuring data accuracy and relevance.
2. **Data Definition**:
  - The Data Definition Language (DDL) is instrumental in managing the structure of tables and indexes within a database.
  - Statements such as CREATE, ALTER, RENAME, DROP, and TRUNCATE are fundamental elements of DDL, enabling the creation, modification, and removal of database objects.
3. **Data Control**:
   - The Data Control Language (DCL) empowers administrators to set permissions for users and user groups, determining whether they have the authority to access and manipulate data.
   - DCL statements are crucial for enforcing security and access control measures.
4. **Transaction**:
  - Transactions play a pivotal role in ensuring the consistency and integrity of a database. A transaction encompasses a series of SQL statements.
  - Once a transaction begins, all these statements are executed, and upon its successful completion, permanent changes are made to the associated tables. - - Transactions are vital for maintaining data reliability.
5. **Procedure**:
  -  Creating reusable methods that contain source code for executing repetitive tasks.
  - Stored procedures streamline database operations by encapsulating logic and promoting code reusability.

### **Compare SQL and NoSQL Databases**

### **Key differences between SQL and NoSQL databases:**

**1. Data Structure 📊**
  - **SQL Databases (Relational)**
   - Store data in structured tables (rows and columns) with a fixed schema.
   - Suitable for structured data with defined relationships.
   - Example: MySQL, PostgreSQL, Oracle, SQL Server

- **NoSQL Databases (Non-Relational)**
   - Store data in various formats: **key-value pairs, documents, wide-column stores, or graphs.**
   - Ideal for semi-structured and unstructured data with flexible schemas.
   - Example: MongoDB (Document), Redis (Key-Value), Cassandra (Column-Family), Neo4j (Graph)

**2. Scalability 🚀**
  - **SQL Databases → Vertically Scalable**
   - Increase system capacity by upgrading hardware (CPU, RAM, SSDs).
   - Limited scalability compared to NoSQL.
  - **NoSQL Databases → Horizontally Scalable**
   - Handle large amounts of data by distributing it across multiple servers (sharding, replication).
   - Preferred for high-traffic applications like social media, e-commerce, and IoT.

**3. Schema Flexibility 🏗️**
  - **SQL Databases → Fixed Schema**
   - Requires predefined tables, columns, data types, and relationships.
   - Schema changes require database migration.
  - **NoSQL Databases → Dynamic Schema**
    - No predefined structure; supports flexible JSON, BSON, or key-value formats.
    - Suitable for applications with evolving data models.

**4. Transactions & ACID Compliance 🔐**
  - **SQL Databases → ACID Compliance**
    - Ensures Atomicity, Consistency, Isolation, and Durability for reliable transactions.
   - Ideal for applications needing financial, banking, and enterprise-grade data integrity.
   - Example
   ```
   BEGIN TRANSACTION;
UPDATE accounts SET balance = balance - 500 WHERE id = 1;
UPDATE accounts SET balance = balance + 500 WHERE id = 2;
COMMIT; -- If both updates succeed
ROLLBACK; -- If any update fails
```
  - **NoSQL Databases → Eventual Consistency**
   - Follows BASE (Basically Available, Soft state, Eventually consistent) principles.
   - Prioritizes speed and availability over strict consistency.
   - Best for real-time applications like social networks, recommendation engines.

**5. Query Language 🔍**
  - **SQL Databases → Use SQL (Structured Query Language)**
    - Standardized language for managing structured data  
    - Example
    ```
    SELECT name, email FROM users WHERE age > 25;
    ```
  - **NoSQL Databases → Use Unstructured Queries (Depends on Database Type)**
    - Document-Based (MongoDB) uses JSON-like queries:
    ```
    { "age": { "$gt": 25 } }
    ```
    - Key-Value (Redis) uses simple GET/SET commands.

    




### **Use Cases Where SQL is Preferred Over NoSQL 🏢**
**1. Banking and Financial Systems 💳**

    - Requires strict ACID compliance to ensure data integrity during transactions.

   - Why SQL?
    - ERP systems require structured data for functions like inventory management, procurement, and finance.
    - SQL's ability to handle complex transactions and maintain relationships between various entities (e.g., suppliers, orders, invoices) is key.
    - Example: SAP, Oracle ERP.

**4. Healthcare Systems 🏥**
  - Why SQL?
    - Patient records and healthcare data must be highly structured, accurate, and adhere to strict compliance standards.
   - ACID compliance ensures the integrity and security of sensitive medical data.
   - Example: Electronic Health Records (EHR) Systems, Hospital Management Systems.

**5. E-Commerce Platforms 🛒**
  - Why SQL?
  - E-commerce platforms often deal with structured data related to products, customers, inventory, orders, and transactions.
  - Relational databases are ideal for handling complex queries such as user purchase history, product recommendations, and inventory management.
  - Example: Magento, Shopify.

###**The History and Evolution of SQL**

####**Origins of SQL**
- SQL (Structured Query Language) was developed in the early 1970s at IBM Research by Donald D. Chamberlin and Raymond F. Boyce.
-  It was initially called SEQUEL (Structured English Query Language) and was designed to manipulate and retrieve data stored in IBM’s System R, one of the first relational database management systems (RDBMS).




#### **Standardization and Adoption**
- In 1979, Relational Software, Inc. (now Oracle Corporation) released the first commercially available SQL-based RDBMS.
- In 1986, SQL was officially standardized by ANSI (American National Standards Institute) as SQL-86, and later by ISO (International Organization for Standardization) in SQL-87.
- Subsequent versions, SQL-89 and SQL-92, introduced improvements such as additional data types, integrity constraints, and enhanced query capabilities.


####**Modern Enhancements and Features**
- SQL:1999 introduced recursive queries, triggers, and procedural extensions (PL/SQL, T-SQL, etc.).
- SQL:2003 brought XML support and introduced window functions for advanced analytics.
- SQL:2011 focused on temporal databases, enabling time-based queries.
- SQL:2016 added support for JSON data, making it more compatible with modern web applications.
- SQL:2019 and beyond introduced big data analytics, machine learning integration, and NoSQL-like capabilities.

####**SQL in the Present and Future**
- Today, SQL remains the dominant language for data management, widely used in data engineering, analytics, and cloud computing.
- Cloud-based SQL services, such as Google BigQuery, Amazon Redshift, and Azure SQL Database, have made SQL even more scalable and accessible.

As data volumes continue to grow, SQL continues to evolve, integrating with AI, machine learning, and big data platforms, ensuring its relevance in the ever-changing data landscape.
